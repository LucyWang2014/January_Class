```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
def pairs(X,target=None,labels=None,cmap=None,figsize=(12,12),spacing=0.1):
    """produce a crude R 'pairs' plot
       X: array of features
       target: classifications (if available)
    """
    #setup graphics device (maybe this belongs outside function)
    plt.figure(figsize=figsize)
    plt.subplots_adjust(wspace=spacing,hspace=spacing)
    dims = X.shape
    n = dims[1] #assumes the input is 2D
    if target is None:
        target=1 #just plot the first color
    if cmap is None:
        cmap = prism # a decent default
    for i in range(0,n-1):
        for j in range(i+1,n):
            plt.subplot(n-1, n-1, (j-1)*3+i+1)
            plt.scatter(X[:, i], X[:, j],c=target,cmap=cmap)
            if j==n-1:
                plt.xlabel(labels[i])
            if i==0:
                plt.ylabel(labels[j])
    plt.show()
```

```python
cmap = mpl.colors.ListedColormap(['b', 'y', 'r'])
pairs(iris.data,Y,iris.feature_names,cmap=cmap)
```

```python
simNagin = pd.read_stata("simNagin.dta")
obs = simNagin.as_matrix()[:,range(12,22)]
```

```python
g = mixture.GMM(covariance_type='diag',n_components=3)
```

```python
%matplotlib inline
g.fit(obs) 
gps = g.predict(obs)
cm = confusion_matrix(simNagin["Group"]-1,gps)
print cm
plt.matshow(cm)
g.bic(obs)
```
