```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
# import some data to play with
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
#all rendering 'internal'
%matplotlib inline
```

```python
cmap = mpl.colors.ListedColormap(['b', 'y', 'r'])
plt.figure(figsize=(12, 12))
plt.subplots_adjust(wspace=.5,hspace=.5)
for i in range(0,3):
    for j in range(i+1,4):
        plt.subplot(3, 3, (j-1)*3+i+1)
        plt.scatter(X[:, i], X[:, j],c=Y,cmap=cmap)
        plt.xlabel(iris.feature_names[i])
        plt.ylabel(iris.feature_names[j])

plt.show()
```

```python
def pairs(X,target=None,labels=None,cmap=None,figsize=(12,12),spacing=0.1):
    """produce a crude R 'pairs' plot
       X: array of features
       target: classifications (if available)
    """
    #setup graphics device (maybe this belongs outside function)
    plt.figure(figsize=figsize)
    plt.subplots_adjust(wspace=spacing,hspace=spacing)
    dims = X.shape
    n = dims[1] #assumes the input is 2D
    if target is None:
        target=1 #just plot the first color
    if cmap is None:
        cmap = prism # a decent default
    for i in range(0,n-1):
        for j in range(i+1,n):
            plt.subplot(n-1, n-1, (j-1)*3+i+1)
            plt.scatter(X[:, i], X[:, j],c=target,cmap=cmap)
            if j==n-1:
                plt.xlabel(labels[i])
            if i==0:
                plt.ylabel(labels[j])
    plt.show()
```

```python
import numpy as np
x = np.linspace(0, 2 * np.pi)
offsets = np.linspace(0, 2*np.pi, 4, endpoint=False)
# Create array with shifted-sine curve along each column
yy = np.transpose([np.sin(x + phi) for phi in offsets])

plt.rc('lines', linewidth=4)
fig, (ax0, ax1)  = plt.subplots(nrows=2)

plt.rc('axes', color_cycle=['r', 'g', 'b', 'y'])
ax0.plot(yy)
ax0.set_title('Set default color cycle to rgby')

ax1.set_color_cycle(['c', 'm', 'y', 'k'])
ax1.plot(yy)
ax1.set_title('Set axes color cycle to cmyk')

# Tweak spacing between subplots to prevent labels from overlapping
plt.subplots_adjust(hspace=0.3)
plt.show()
```
